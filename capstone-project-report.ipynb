# Dubai Neighborhood Analysis

## Introduction


Emirate of Dubai is one of the seven emirates of United Arab Emirates (U.A.E.). With a population of 3.3 million, it is the most populated emirate of UAE. Dubai annual GDP is around USD 108 billions with economy focused around Tourism, Retail, Real estate investments, and Commodities trading.

For administrative purpose, Dubai is divided into 226 communities (neighborhoods) which are grouped in 9 sectors. Coastal communities are densely populated and are the major centers of economic activities, with spare to no population towards the interior desert communities.

As Capstone project for IBM Data Science certification, my objective is to analyze the various neighborhoods of Dubai for their economic and social activities. This analysis would provide a neighbourhood comparision for their business activities and population growth/de-growth. With the sourced data as proxy for economic/social activities, I aim to build a Dubai Neighborhood Index that will rank the communities.

Such insights would be useful to business startups which can help them identify socially intense areas, areas of high population growth, and business activities hotspots.

## Data Description


Data on population numbers per neighborhood was fetch from [Dubai Statistics Center](http://https://www.dsc.gov.ae/) which publish annual popuation stats. The latest available years are for 2019 and 2018 which are used in this report.

Geometrics for neighborhood boundaries were available from Dubai Municipality as KML file, which was converted to geojson format. Land Area was calculated based on polygons available in this file. I have merged the population stats in geojson file for calculating the neighborhood population density.

Social data on business activities is fetch through FourSquare Venue APIs which in this analysis utilizes the checkins as a proxy indicator for business intensity.
